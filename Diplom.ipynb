### 7. Проверить гипотезу, что сотрудники с высоким окладом проводят на работе больше времени, чем сотрудники с низким окладом.

```python
high_salary = HR[(HR['salary']=='high')]
high_salary[['average_montly_hours','time_spend_company']].sum()
```

```python
low_salary = HR[(HR['salary']=='low')]
low_salary[['average_montly_hours','time_spend_company']].sum()
```

```python
# сравниваю показатели
high_salary[['average_montly_hours','time_spend_company']].sum() > low_salary[['average_montly_hours','time_spend_company']].sum()
```

### Вывод: данная гипотеза не верна! 
- сотрудники с низким окладом на работе, проводят больше времени, чем сотрудники с высоким окладом, и к тому же стаж работы низкооплачиваемых сотрудников выше(они больше лет провели на одном и том же рабочем месте) 

### 8.Рассчитать следующие показатели среди уволившихся и неуволившихся сотрудников (по отдельности): 
- Доля сотрудников с повышением за последние 5 лет
- Средняя степень удовлетворенности
- Среднее количество проектов

```python
# уволенные сотрудники и получивышие повышение за последние 5 лет
HR[(HR['left']==1)
   & (HR['promotion_last_5years']==1)      
].department.count() 
```

```python
# неуволенные сотрудники и получивышие повышение за последние 5 лет
HR[(HR['left']==0)
   & (HR['promotion_last_5years']==1)
].department.count()
```

```python
left1 = HR[HR['left']==1] # уволенные сотрудники
left1.department.count()
```

```python
left0 = HR[HR['left']==0] # неуволенные сотрудники
left0.department.count()
```

```python
print(left1['satisfaction_level'].mean(),'Средняя степень удовлетворенности уволенных сотрудников')
print(left0['satisfaction_level'].mean(),'Средняя степень удовлетворенности неуволенных сотрудников')
# для расчета среднего количества проектов для сотрудника корректней будет испльзовать медиану
print(left1['number_project'].median(),'Среднее количество проектов для уволенных людей')
print(left0['number_project'].median(),'Среднее количество проектов для неуволенных людей')
```
